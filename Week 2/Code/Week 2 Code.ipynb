```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def g_1(x):
    g1 = np.square(x)-2
    return g1
def g_2(x):
    g2 = np.sqrt(x+2)
    return g2
```

```python
x = np.linspace(0,3,num = 150)
plt.plot(x,g_1(x),label = "$g_1(x)$")
plt.plot(x,g_2(x),label = "$g_2(x)$")
plt.plot(x,x,label = "$y = x$")
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.legend()
plt.show()
```

```python
r = 5
maxniter = 10
r_save = np.empty(maxniter+1)
r_save[0] = r
for i in range(maxniter):
    r = g_2(r)
    r_save[i+1] = r
```

```python
plt.semilogy(range(maxniter+1),np.abs(r_save-2),'o',label = "Error")
plt.semilogy(range(maxniter+1),np.power(1/4,range(maxniter+1)),'k-',label = "(1/4)^k")
plt.xlabel("Steps")
plt.ylabel("Error")
plt.legend()
plt.show()
```
